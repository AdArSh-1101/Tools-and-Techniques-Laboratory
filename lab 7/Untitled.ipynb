```python
import cv2
```

```python
img=cv2.imread('b.png',1)
```

```python
import matplotlib.pyplot as plt
```

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
cv2.imshow('Original Image', img)
cv2.waitKey(0)
```

```python
rimg = cv2.resize(img, (800, 800))
```

```python
cv2.imshow('Original Image', rimg)
cv2.waitKey(0)
image = cv2.cvtColor(rimg, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
rows, cols = rimg.shape[:2]
rm = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
rotated_image = cv2.warpAffine(rimg, rm, (cols, rows))
```

```python
cv2.imshow('Original Image', rotated_image)
cv2.waitKey(0)
image = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
flipped_image = cv2.flip(rimg, 1)

cv2.imshow('Flipped Image', flipped_image)
cv2.waitKey(0)
image = cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(rimg, cv2.COLOR_BGR2GRAY)
cv2.imshow('Original Image', gray_image)
cv2.waitKey(0)
image = cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
edges = cv2.Canny(gray_image, 100, 200)
cv2.imshow('Edge Detected Image', edges)
cv2.waitKey(0)
image = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()

contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
cv2.drawContours(rimg, contours, -1, (0, 255, 0), 2)

# Display the image with contours
cv2.imshow('Image with Contours', rimg)
cv2.waitKey(0)
image = cv2.cvtColor(rimg, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```
